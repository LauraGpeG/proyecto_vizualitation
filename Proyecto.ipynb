```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
from jupyter_dash import JupyterDash # Versión de dash para notebooks
import dash_core_components as dcc # Componentes HTML preconstruidos para dashboards
import dash_html_components as html # Componentes HTML nativos
from dash.dependencies import Input, Output # Clases Input y Output
```

```python
df = pd.read_csv("/content/drive/MyDrive/CO2_emission.csv")
```

```python
df.info()
```

```python
df.head()
```

Cambiamos la orientación de los datos y el tipo de dato en la columna Year de object a float64

```python
df = (df.dropna()).drop('Indicator Name', axis=1)
```

```python
df2 = df.set_index(['Country Name', 'country_code', 'Region']).stack()
df2 = df2.to_frame('CO2')
df2.index = df2.index.set_names('Year', level=3)
df2 = df2.reset_index()
df2['Year'] = df2['Year'].apply(np.float64)
df2.info()
```

```python
df19 = df2.query("Year == 2019").groupby('Region').sum()
df19.head()
```

```python
fig1=px.scatter(
    df2, 
    x='Year', 
    y='CO2',
    size='CO2', 
    color='Region',
    hover_name='Country Name', 
    log_x=True, 
    size_max=60
    )
fig1.show()
```

```python
df3 = df2.sort_values('CO2',ascending=True)
df3.head()
```

```python
fig2 = px.bar(df3, x='CO2', y='Country Name', color='CO2', title="Emisiones de CO2 por país")
fig2.show()
```

```python

px.scatter_geo(
    df2, 
    locations='country_code', 
    color='CO2',
    color_continuous_scale=[(0.00, "green"),(1.00, "red")],
    hover_name='Country Name', 
    size='CO2',
    animation_frame='Year', 
    projection='natural earth',
    labels=dict(color='Region'),
    title='Variación de CO2 entre 1990-2019'
)
```

```python
 fig = px.choropleth(
        data_frame=df2,
        locationmode='country names',
        locations='country_code',
        scope='world',
        color='CO2',
        hover_data=['Region','CO2','Year'],
        color_continuous_scale=px.colors.sequential.YlOrRd,
        template='plotly_dark'
    )
 fig.show()
```

```python

app=JupyterDash(__name__)

app.layout = html.Div([
    
    html.H1("Emisiones de CO2 alrededor del mundo", style={'text-align':'center'}),

    dcc.Graph(id='graphic-CO2'),
    dcc.RangeSlider(
        df2['Year'].min(),
        df2['Year'].max(),
        step=None,
        value=df2['Year'].min(),
        marks={str(year): str(year) for year in df2['Year'].unique()},
        id='year-slider'
    )
])

@app.callback(
    Output('graphic-CO2', 'figure'),
    Input('year-slider', 'value'))
def update_graph(selected_year):
    filtered_df = df2[df2.year == selected_year]
    
    fig = px.choropleth(
        data_frame=df2,
        locationmode='country names',
        locations='country_code',
        scope='world',
        color='CO2',
        hover_data=['Region','CO2','Year'],
        color_continuous_scale=px.colors.sequential.YlOrRd,
        template='plotly_dark'
    )
    
    return fig


if __name__ == '__main__':
    app.run_server(debug=True, mode='inline')
```
